```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("Accidents.csv")
df = df.iloc[:3000]
```

```python
df.dtypes
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x= 'Weather_conditions',palette='flare',hue='Day')
plt.title("Perbandingan Kondisi Cuaca Tiap Hari")
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik Batang yang menggunakan seaborn yang menjelaskan tentang jumlah data Kondisi Cuaca pada hari terjadi Kecelakaan dari dataset yang 
berjumlah 3000 data dengan 
xlabel: Kondisi dan ylabel: kuantitas

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x= 'Day',palette='flare')
plt.title("Perbandingan Jumlah Hari Pada Data")
plt.xlabel('Nama Tiap Hari')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik Batang yang menggunakan seaborn yang menjelaskan tentang jumlah data Tiap-Tiap Hari pada dataset yang berjumlah 3000 data dengan 
xlabel: Nama Tiap Hari dan ylabel: kuantitas

```python
data_pie = df["Day"].value_counts().rename_axis("Day").reset_index(name="id_count")
```

```python
plt.figure(figsize=(8,8))
plt.pie(data_pie["id_count"],labels=data_pie["Day"],startangle=120,autopct='%.1f%%')
plt.xlabel('Nama Tiap Hari')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik Pie ini menampilkan data dalam bentuk Persenan dari Jumlah Tiap-Tiap hari yang berada pada dataset yang totalnya berjumlah 3000

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, cmap='Blues')
ax.set_title('Bicycle Accidents in Great Britain (1979 to 2018)')
```

```python
Grafik HeatMap ini merupakan visualisasi dari dataset Bicycle Accidents in Great Britain (1979 to 2018), 
yang menjelaskan relasi antar 3 variabel yang terdapat di dataset
mulai dari Jumlah Kendaraan, Jumlah Korban, dan Batas Kecepatan.
```
